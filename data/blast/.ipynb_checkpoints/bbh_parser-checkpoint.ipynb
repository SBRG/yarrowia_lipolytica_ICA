```python
import pandas as pd
from Bio import SeqIO
```

```python
bbh_table = pd.read_csv('bbh/GCF_000146045.2_R64_protein_vs_GCF_000002525.2_ASM252v1_protein_parsed.csv',index_col='Unnamed: 0')
yarrowia_annotations = pd.read_csv('tables/yarrowia_NCBI_table.csv',index_col='#Name')
sacc_annotations = pd.read_csv('tables/scerevisiae_NCBI_table.csv',index_col='#Name')
```

```python
blast_dataframe = pd.DataFrame(columns=['gene AID', 'gene name','subject AID','subject name',
                                       'gene product AID','subject product AID','gene product name','subject product name','COV'])
```

```python
for i in range (0,bbh_table.shape[0]):
    gene_aid = list(bbh_table['gene'])[i]
    subject_aid = list(bbh_table['subject'])[i]
    subject_name = yarrowia_annotations.loc[yarrowia_annotations['Protein product']==subject_aid]['Locus tag'][0]
    gene_name = sacc_annotations.loc[sacc_annotations['Protein product']==gene_aid]['Locus tag'][0]
    gene_product = sacc_annotations.loc[sacc_annotations['Protein product']==gene_aid]['Protein product'][0]
    subject_product = yarrowia_annotations.loc[yarrowia_annotations['Protein product']==subject_aid]['Protein product'][0]
    gene_product_name = sacc_annotations.loc[sacc_annotations['Protein product']==gene_aid]['Protein Name'][0]
    subject_product_name = yarrowia_annotations.loc[yarrowia_annotations['Protein product']==subject_aid]['Protein Name'][0]
    cov = list(bbh_table['COV'])[i]
    blast_dataframe.loc[len(blast_dataframe.index)] = [gene_aid, gene_name, subject_aid,subject_name,
                                                      gene_product,subject_product,gene_product_name,
                                                      subject_product_name,cov]
```

```python
sacc_gene_set = pd.read_table('tables/SCerevisiaeGeneSet.tsv')
sacc_gene_set
```

```python
sacc_TRN = pd.read_csv('tables/SCerevisiaeTRN.tsv')
sacc_TRN
```

```python
TFs = []
TFs_genes = []

for string_element in sacc_TRN['TF/Gene']:
    tf_string = ''
    tf_gene_string = ''
    for char in string_element:
        if(char != ';'):
            tf_string = tf_string + char
        if(char ==';'):
            index = string_element.index(';')
            tf_gene_string = string_element[index+1:]
            break
    TFs.append(tf_string)
    TFs_genes.append(tf_gene_string)
            
```

```python
sacc_TRN_dataframe = pd.DataFrame(columns=['regulator','gene_name'])
sacc_TRN_dataframe['regulator']=TFs
sacc_TRN_dataframe['gene_name'] = TFs_genes
gene_id = []

for gene in sacc_TRN_dataframe['gene_name']:
    gene_name = list(sacc_gene_set[sacc_gene_set['Gene name']==gene]['Gene ID'])
    if(len(gene_name)):
        gene_id.append(list(sacc_gene_set[sacc_gene_set['Gene name']==gene]['Gene ID']).pop())
    else:
        gene_id.append('None')

        
sacc_TRN_dataframe['sacc_gene_id']=gene_id

yarrowia_gene_match = []
for gene_id in sacc_TRN_dataframe['sacc_gene_id']:
    blast_hits = blast_dataframe[blast_dataframe['gene name']==gene_id]
    yarrowia_gene_match = []
    if(blast_hits.shape[0]):
        yarrowia_gene_match.append(blast_hits['subject name'].tolist().pop())
    
sacc_TRN_dataframe['gene_id']=yarrowia_gene_match
sacc_TRN_dataframe
```
